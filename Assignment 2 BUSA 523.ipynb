```python
# creating student class
class Student:
    # init method to take values
    def __init__(self,first_name,last_name,country,overall_grade,credits_completed):
        self.first_name  = first_name
        self.last_name = last_name
        self.country = country
        self.overall_grade = overall_grade
        self.credits_completed = credits_completed

    # creating a progress method
    def progress(self):
        if self.overall_grade > 3.0 and self.credits_completed > 20:
            print("Good")
        elif self.overall_grade > 2.5 and self.credits_completed <= 20 and self.credits_completed >= 10:
            print("Average")
        elif self.overall_grade <= 2.5:
            print("Below Average")

# example 1
print("example1:-",end=" ")
stud1 = Student('ram','kumar','US',3.5,25);
stud1.progress();

# example 2
print("example2:-",end=" ")
stud2 = Student('ravi','kumar','Canada',2.8,18);
stud2.progress();
```

```python
#User input 
myList = list(map(int, input("Input: ").split(",")))
#Declare lists 
positive = []
negative = []
#Iterate through the list 
for i in myList:
	#Check if negative
	if i<0:
		#Append to negative
		negative.append(i)
	elif i>0:
		#Append to positive list
		positive.append(i)
#Output Result
print("Output:",negative,positive)
```

```python

score = int(input("Enter players's score between 15 and 21 "))

answer = [] # empty list for storing answer

for i in range(1,12):
    for j in range(1,i+1):
        if j==11:
            break
        if i+j > score:
            answer.append(tuple((i,j)))

print(answer)
```
