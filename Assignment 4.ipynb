```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.ticker import AutoMinorLocator
from matplotlib import gridspec
```

```python
df=pd.read_csv('data.csv')
df
df['Date']=pd.to_datetime(df['Date'], utc=True)
```

```python
fig, ax = plt.subplots(1, figsize=(14,6))
n, bins, patches = plt.hist(df['CPI'])
# define minor ticks and draw a grid with them
minor_locator = AutoMinorLocator(2)
plt.gca().xaxis.set_minor_locator(minor_locator)
plt.grid(which='minor', color='Black', lw = 0.5)
# x ticks
xticks = [(bins[idx+1] + value)/2 for idx, value in enumerate(bins[:-1])]
xticks_labels = [ "{:.2f}\nto\n{:.2f}".format(value, bins[idx+1]) for idx, value in enumerate(bins[:-1])]
plt.xticks(xticks, labels = xticks_labels)
plt.title('Histogram of CPI', loc = 'left', fontsize = 18)
```

```python
plt.figure(figsize = (14,7))
x = df.Date
y = df.CPI 
plt.bar(x, y, width= 0.7, align='center',color='Red', edgecolor = 'Green')
plt.legend(labels = ['CPI'])
plt.title("Bar plot representing the CPI against Temperature")
plt.xlabel('Date')
plt.ylabel('CPI')
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.plot(y,)
plt.legend(labels = ['Temperature'])
plt.title("Line plot representing Temperature")
plt.show()
```

```python
plt.figure(figsize = (14,7))
plt.legend(labels = ['Fuel_Price'])
plt.scatter(df['Date'], df['Fuel_Price'])
plt.xlabel('Date')
plt.ylabel('Fuel_Price')
plt.title("Scatter plot representing  Fuel_Price")
plt.show()
```
